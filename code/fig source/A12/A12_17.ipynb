```python
from transformers import AutoTokenizer
from transformers import LlamaForCausalLM
import torch
```

```python
device = torch.device("cuda")
model_dir = "/root/code/LLM/LLaMa"
model = LlamaForCausalLM.from_pretrained(model_dir, torch_dtype=torch.float16, device_map="auto", output_attentions=True)
tokenizer = AutoTokenizer.from_pretrained(model_dir, use_fast=False)
```

```python
inputs = tokenizer("John's father is Mike. Mike's father is Tom. Who's the grandfather of John?", return_tensors="pt").to(device)
input_ids = inputs['input_ids']

attention = model(input_ids)[-1]
```

```python
from bertviz import head_view
def show_head_view(model, tokenizer, sentence_a, sentence_b=None, layer=None, heads=None):
    inputs = tokenizer("My name is John. What is my name? Your name is John", return_tensors="pt").to(device)
    input_ids = inputs['input_ids']

    attention = model(input_ids)[-1]
    sentence_b_start = None
    input_id_list = input_ids[0].tolist() # Batch index 0
    tokens = tokenizer.convert_ids_to_tokens(input_id_list)
    head_view(attention, tokens, sentence_b_start, layer=layer, heads=heads)
show_head_view(model, tokenizer, 'sentence_a')
```
