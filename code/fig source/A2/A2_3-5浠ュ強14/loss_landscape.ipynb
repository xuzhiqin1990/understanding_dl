$$-20 * e^{-0.2 * \sqrt{0.5 * (x^2 + y^2)}} - e^{0.5 * (cos(2*pi*x) + cos(2*pi*y))} + e + 20$$

```python
def ackley(x, y):
    return -20 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2))) - np.exp(0.5 * (np.cos(2*np.pi*x) + np.cos(2*np.pi*y))) + np.e + 20

# (-5,5)
x = np.linspace(-1.5, 1.5, 100)
y = np.linspace(-1.5, 1.5, 100)
X, Y = np.meshgrid(x, y)

# Z
Z = ackley(X, Y)

# 3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# 
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# 
ax.view_init(elev=15, azim=60)

# 
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.grid(False)

# 
ax.axis('off')


# 
plt.tight_layout()
plt.show()
```

```python
def peaks_and_valleys(x, y):
    return 3*np.exp(-((x-1)**2 + (y-1)**2) / 0.7**2) - 2*np.exp(-((x+1)**2 + (y+1)**2) / 0.5**2) #+ 2*np.exp(-((x+1)**2 + (y-1)**2) / 0.3**2)

# (-3,3)
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Z
Z = peaks_and_valleys(X, Y)

# 3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# 
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# 
ax.view_init(elev=10, azim=-95)

# 
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.grid(False)

# 
ax.axis('off')


# 
plt.tight_layout()
plt.show()
```
